```python
import fft
import h5py
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
with h5py.File('../dmft_delta/beta_50_gamma_bd_0.3') as f:
    gf = f['gf'][:]
gf = np.concatenate((np.conj(gf[::-1]), gf))
```

```python
plt.plot(gf.imag)
```

```python
gt = fft.inverse_fourier(50.0, gf, n_points=10000)
gw = fft.fourier(50., gt, 256)
```

```python
plt.plot(gt)
```

```python
plt.plot(gw.imag,'-o')
plt.plot(gf.imag)
```

```python
plt.plot(gw.real,'-o')
plt.plot(gf.real)
```
